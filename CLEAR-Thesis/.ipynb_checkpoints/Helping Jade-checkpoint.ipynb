```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure
import matplotlib
from scipy.optimize import curve_fit

x_bins=60;y_bins=60;z_bins=250
sim=pd.read_csv("D:Desktop/helpingjade/results/dosetomedium_helpingjade.csv",delimiter= ',',comment='#',names=['x','y','z','dose'])
sim= sim.dose.values.reshape(x_bins, y_bins, z_bins)
sim=sim.T
sim=sim/(1.60218e-19)*(100*10**-12)

```

```python
filmx=[]
filmy=[]
for i in np.arange(1,7):
    filmx.append(np.load("D:Desktop/helpingjade/film_x_"+str(i)+".npy"))
    filmy.append(np.load("D:Desktop/helpingjade/film_y_"+str(i)+".npy"))

def gaus(x,a,x0,sigma):
    return a*np.exp(-(x-x0)**2/(2*sigma**2))

def gauss_fit(xdata,ydata):
    mean = sum(xdata * ydata) / sum(ydata)
    sigma = np.sqrt(sum(ydata * (xdata - mean)**2) / sum(ydata))
    popt,pcov = curve_fit(gaus,xdata,ydata,p0=[max(ydata), mean, sigma])
    #plt.plot(xdata,gaus(xdata,*popt),label='Gaussian fit',lw=1)
    return popt[-1],np.sqrt(np.diag(pcov))[-1]


for i in np.arange(1,6):
    figure(i,figsize=(6,6))
    plt.plot(np.arange(0,len(filmx[i-1]))*0.085,filmx[i-1],label='film_x')
    plt.plot(np.arange(0,len(sim[i*36,30]))*0.1+1.1,sim[i*36,30],label='Sim')
    plt.legend()

for i in np.arange(1,6):
    figure(i*10,figsize=(6,6))
    plt.plot(np.arange(0,len(filmy[i-1]))*0.085,filmy[i-1],label='film_y')
    plt.plot(np.arange(0,len(sim[i*36,:,30]))*0.1+1.1,sim[i*36,:,30],label='Sim')
    
    plt.legend()

for i in np.arange(1,6):
    print(gauss_fit(np.arange(0,len(filmx[i-1]))*0.085,filmx[i-1])[0],gauss_fit(np.arange(0,len(sim[i*36,30]))*0.1+1.1,sim[i*36,30])[0])
    print(gauss_fit(np.arange(0,len(filmy[i-1]))*0.085,filmy[i-1])[0],gauss_fit(np.arange(0,len(sim[i*36,:,30]))*0.1+1.1,sim[i*36,:,30])[0])
```
