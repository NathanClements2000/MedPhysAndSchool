```python
import numpy as np
import scipy as sp
from scipy.integrate import odeint
from matplotlib import pyplot as plt
from matplotlib.pyplot import figure
m=1
def scatang(E,s):
    #initial conds
    r=[50]
    theta=[np.pi-np.arcsin(s/r[0])]
    rdot=[-np.sqrt(2*E/m-2/(m*r[0]))] #from E=T+V
    
    #time evolution
    tstep=0.001
    t=np.arange(0,100,tstep)
    for i in t:
        r.append(r[-1]+rdot[-1]*tstep)
        theta.append(theta[-1]+s*np.sqrt(2*m*E)/(m*r[-1]**2)*tstep)
        rdot.append(rdot[-1]+ (2*E*s**2/(m*r[-1]**3)+1/r[-1]**2)*tstep)
    return(theta[-1])
# test=scatang(100,0.1)

# figure(1,figsize=(8,8))
# plt.plot(test[0]*np.cos(test[2]),test[0]*np.sin(test[2]))
# plt.grid(c='lightgray')

#E=10
E10ang=[]
#calculate scattering angles for specific E using range of s from 0.01 to 1
srange=np.linspace(0.01,1,100)
for j in srange:
    E10ang.append(scatang(10,j))
#since linear, slope is simply:
ds_dtheta=np.abs((srange[-1]-srange[0])/(E10ang[-1]-E10ang[0]))
sigma10=srange*ds_dtheta/np.sin(E10ang)
figure(1,figsize=(8,8))
plt.plot(E10ang,-sigma10,label='Calculated result')
plt.xlabel('Scattering angle (rad)')
plt.ylabel('Sigma')
plt.legend()

#Rutherford
def sigma_R(E,angle):
    E=np.array(E)
    angle=np.array(angle)
    return 0.25*(0.5/E)**2*(1/np.sin(angle/2)**4)

figure(1,figsize=(8,8))
plt.plot(E10ang,sigma_R(10,E10ang),label="Rutherford's result")
plt.title('E=10')
plt.grid(c='lightgray')
plt.legend()

#E=1
E1ang=[]
#calculate scattering angles for specific E using range of s from 0.01 to 1
srange=np.linspace(0.01,1,100)
for j in srange:
    E1ang.append(scatang(1,j))
#since linear, slope is simply:
ds_dtheta=np.abs((srange[-1]-srange[0])/(E1ang[-1]-E1ang[0]))
sigma1=srange*ds_dtheta/np.sin(E1ang)
figure(4,figsize=(8,8))
plt.plot(E1ang,-sigma1,label='Calculated result')
plt.plot(E1ang,sigma_R(1,E1ang),label="Rutherford's result")
plt.xlabel('Scattering angle (rad)')
plt.ylabel('Sigma')
plt.title('E=1')
plt.grid(c='lightgray')
plt.legend()


#E=50
E50ang=[]
#calculate scattering angles for specific E using range of s from 0.01 to 1
srange=np.linspace(0.01,1,100)
for j in srange:
    E50ang.append(scatang(50,j))
#since linear, slope is simply:
ds_dtheta=np.abs((srange[-1]-srange[0])/(E50ang[-1]-E50ang[0]))
sigma50=srange*ds_dtheta/np.sin(E50ang)
figure(2,figsize=(8,8))
plt.plot(E50ang,-sigma50,label='Calculated result')
plt.plot(E50ang,sigma_R(50,E50ang),label="Rutherford's result")
plt.xlabel('Scattering angle (rad)')
plt.ylabel('Sigma')
plt.title('E=50')
plt.grid(c='lightgray')
plt.legend()

#E=2
E2ang=[]
#calculate scattering angles for specific E using range of s from 0.01 to 1
srange=np.linspace(0.01,1,100)
for j in srange:
    E2ang.append(scatang(2,j))
#since linear, slope is simply:
ds_dtheta=np.abs((srange[-1]-srange[0])/(E2ang[-1]-E2ang[0]))
sigma2=srange*ds_dtheta/np.sin(E2ang)
figure(3,figsize=(8,8))
plt.plot(E2ang,-sigma2,label='Calculated result')
plt.plot(E2ang,sigma_R(2,E2ang),label="Rutherford's result")
plt.xlabel('Scattering angle (rad)')
plt.ylabel('Sigma')
plt.title('E=2')
plt.grid(c='lightgray')
plt.legend()


print('This was as close as I could get, the discrepancy between the two results are likely a result of numerical precision')
```
