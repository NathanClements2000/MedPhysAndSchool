```python
import numpy as np
import scipy as sp
from scipy.integrate import odeint
from matplotlib import pyplot as plt
from matplotlib.pyplot import figure
#define constants
m=1
k=-1
r0=1
#make use of theta'=l/(mr^2), r''=l^2/(m^2 r^3) - k a r^(a-1), and r_c=(l^2/(mka))^(1/(a+2))

def f(a,l):
    #initial conds
    l0=np.sqrt(m*k*a*r0**(2+a))
    r=[r0]
    rdot=[0]
    theta=[0]
    thetadot=[0]
    
    #time evolution
    tstep=0.001
    t=np.arange(0,10,tstep)
    for i in t:
        r.append(r[-1]+rdot[-1]*tstep)
        theta.append(theta[-1]+l*l0/(m*r[-1]**2)*tstep)
        rdot.append(rdot[-1]+ ((l*l0)**2/(m**2*r[-1]**3) - k*a*r[-1]**(a-1))*tstep)
    #find index of r where it returns to maximum r
    h=[]
    for i in np.arange(0,len(r)-1):
        if (r[i]>r[i+1]) & (r[i]>r[i-1]):
            h.append(i)
    return(theta[h[0]]-2*np.pi)

# l=0.7
result=[]
a_range=np.linspace(-1.2,-0.8,100)
for i in a_range:
    result.append(f(i,0.7))
figure(1,figsize=(8,8))
plt.plot(a_range,result,label='l=0.7')
plt.grid(c='lightgray')
plt.xlabel('a')
plt.ylabel('theta-2pi')
# l=0.75
result=[]
a_range=np.linspace(-1.2,-0.8,100)
for i in a_range:
    result.append(f(i,0.75))
figure(1,figsize=(8,8))
plt.plot(a_range,result,label='l=0.7')
plt.grid(c='lightgray')
plt.xlabel('a')
plt.ylabel('theta-2pi')
# l=0.8
result=[]
a_range=np.linspace(-1.2,-0.8,100)
for i in a_range:
    result.append(f(i,0.8))
figure(1,figsize=(8,8))
plt.plot(a_range,result,label='l=0.8')
# l=0.8
result=[]
a_range=np.linspace(-1.2,-0.8,100)
for i in a_range:
    result.append(f(i,0.85))
figure(1,figsize=(8,8))
plt.plot(a_range,result,label='l=0.8')
# l=0.9
result=[]
a_range=np.linspace(-1.2,-0.8,100)
for i in a_range:
    result.append(f(i,0.9))
figure(1)
plt.plot(a_range,result,label='l=0.9')

# l=0.95
result=[]
a_range=np.linspace(-1.2,-0.8,100)
for i in a_range:
    result.append(f(i,0.95))
figure(1)
plt.plot(a_range,result,label='l=0.95')

# l=0.99
result=[]
a_range=np.linspace(-1.2,-0.8,100)
for i in a_range:
    result.append(f(i,0.99))
figure(1)
plt.plot(a_range,result,label='l=0.99')

#plot e) from analytical
plt.plot(a_range,-2*np.pi*(1-1/(np.sqrt(2-np.abs(a_range)))),marker='o',ms=3,label='Analytical prediction',markevery=5,c='limegreen')
plt.legend()
print('As l -> 1 we approach the analytical solution, great')
```
