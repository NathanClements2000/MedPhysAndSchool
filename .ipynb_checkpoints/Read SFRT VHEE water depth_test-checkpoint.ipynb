```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure
import matplotlib
import scipy
from scipy.signal import find_peaks
import pandas as pd

all_data={}
for j in [150,250]:
    my_array=np.zeros((200,200,100))
    data03=pd.read_csv("D:Desktop/SFRT-Topas/vhee/vhee_patient/depth_test/results/dosetomedium_water_depthtest_"+str(j)+"MV__1.csv",delimiter= ',',comment='#',names=['x','y','z','dose'])
    for i in np.arange(2,6):
        data1=pd.read_csv("D:Desktop/SFRT-Topas/vhee/vhee_patient/depth_test/results/dosetomedium_water_depthtest_"+str(j)+"MV__"+str(i)+".csv",delimiter= ',',comment='#',names=['x','y','z','dose'])
        data03.dose=data03.dose+ data1.dose
    data03.dose=(data03.dose/5e6)/(1.60218e-19)*0.75e-6;
    for i in np.arange(0,len(data03.x)):
        my_array[data03.x[i]][data03.y[i]][data03.z[i]]=data03.dose[i]
    all_data[j]=np.array(my_array)

```

```python
leftedge=43
rightedge=156
xslice=99
maxdose=150
zentry=0

y= np.arange(0,200)*0.25
x= np.arange(0,200)*0.25
z= np.arange(0,100)

DOC_energy=[]
pvdr_energy=[]

for i in [150,250]:
    fig, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize=(13.5,5))
    im1=ax3.imshow(all_data[i][100].T, aspect='auto',extent=(0,200*0.25,0,100),cmap='magma',origin='lower',vmax=maxdose)
    ax3.set_ylabel('z (mm)')
    ax3.set_xlabel('y (mm)')
    ax3.set_xlim((leftedge*0.25,rightedge*0.25))
    ax3.set_ylim((0,60))
    fig.colorbar(im1,label='Gy/s')
    ax3.set_title('2D profile')
    
    
    ax1.plot(y,all_data[i][xslice,:,zentry],lw=2)
    ax1.set_ylim((0,maxdose))
    ax1.set_xlim((leftedge*0.25,rightedge*0.25))
    ax1.set_ylabel('Gy/s')
    ax1.set_xlabel('y (mm)')
    ax1.grid()
    
    
    peaks= find_peaks(all_data[i][xslice,leftedge:rightedge,zentry],height=0.5e8*0.75e-6,distance=4)[0]
    valleys= find_peaks(-all_data[i][xslice,leftedge:rightedge,zentry],distance=4)[0]
    ax1.scatter(y[leftedge+peaks],all_data[i][xslice,leftedge+peaks,zentry],marker='x',c='red')
    ax1.scatter(y[leftedge+valleys],all_data[i][xslice,leftedge+valleys,zentry],marker='x',c='green')
    ax1.set_title('1D Surface profile')

    peaks= find_peaks(all_data[i][xslice,leftedge:rightedge,zentry],height=0.5e8*0.75e-6,distance=4)[0]+leftedge
    peak_locs=[[j,k] for j in peaks for k in peaks]
    valleys= find_peaks(-all_data[i][xslice,46:rightedge,zentry],distance=4)[0]+46
    valley_locs=[[j,k] for j in valleys for k in valleys]
    mean_peaks=[]
    peak_error=[]
    mean_valleys=[]
    valley_error=[]
    for j in np.arange(0,100):
            mean_peaks.append(np.mean([all_data[i][k,m,j] for k,m in peak_locs]))
            peak_error.append(np.std([all_data[i][k,m,j] for k,m in peak_locs])/np.sqrt(len([all_data[i][k,m,j] for k,m in peak_locs])))           
            mean_valleys.append(np.mean([all_data[i][k,m,j] for k,m in valley_locs]))
            valley_error.append(np.std([all_data[i][k,m,j] for k,m in valley_locs])/np.sqrt(len([all_data[i][k,m,j] for k,m in valley_locs]))) 
    mean_peaks=np.array(mean_peaks);peak_error=np.array(peak_error);mean_valleys=np.array(mean_valleys);valley_error=np.array(valley_error)
    pvdr_depth=[mean_peaks[k]/mean_valleys[k] for k in np.arange(0,len(mean_valleys))]
    pvdr_error=np.sqrt((np.array(peak_error)/np.array(mean_peaks))**2+(np.array(valley_error)/np.array(mean_valleys))**2)*pvdr_depth

    
    
    ax2.plot(z,pvdr_depth,label=str(i)+'MV',alpha=1,lw=3)
    ax2.fill_between(z, pvdr_depth-pvdr_error, pvdr_depth+pvdr_error,alpha=0.4)
    ax2.grid()
    ax2.set_ylim((0,41))
    ax2.set_title("PVDR vs depth")
    ax2.set_ylabel('PVDR')
    ax2.set_xlabel('z (mm)')
    DOC= [w for w in np.arange(0,len(pvdr_depth)) if np.max(pvdr_depth[w]-pvdr_error[w])<=1.1][0]
    ax2.vlines(DOC,0,100,colors='red',lw=1)
    ax2.set_ylim(0,60)
    ax2.set_xlim(0,60)
    print(pvdr_depth_avg[0],pvdr_error[0])
    fig.tight_layout()
    if i==150:
        peakPDD150=mean_peaks
    figure(3,figsize=(8,8))
    plt.plot(z,mean_peaks,label=str(i)+' MeV',lw=2)
    plt.fill_between(z,np.array(mean_peaks)-np.array(peak_error),np.array(mean_peaks)+np.array(peak_error),alpha=0.4)
    plt.grid()
    plt.xlim(0,100)
    plt.ylim(0,200)
    plt.xlabel('Depth (mm)')
    plt.ylabel('Dose rate (Gy/s)')
    plt.legend()
    plt.title('Mini-GRID')
    plt.grid(c='lightgray')
np.save("d:Desktop/SFRT-Topas/vhee/vhee_patient/depth_test/meanpeaks_250.npy",mean_peaks)
np.save("d:Desktop/SFRT-Topas/vhee/vhee_patient/depth_test/meanvalleys_250.npy",mean_valleys)


```
