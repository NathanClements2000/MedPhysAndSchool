```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure
import matplotlib
from scipy.interpolate import interp1d

font = {'family' : 'Arial',
        'weight' : 'normal',
        'size'   : 18}

matplotlib.rc('font', **font)



PDD100MV=np.load('D:Desktop/SFRT-Topas/vhee/vhee_bestopt/results/PDD100MV.npy')
pvdr100MV=np.load('D:Desktop/SFRT-Topas/vhee/vhee_bestopt/results/pvdr100MV.npy')
peak100MV=np.load('D:Desktop/SFRT-Topas/vhee/vhee_bestopt/results/peak100MV.npy')

PDD150MV=np.load('D:Desktop/SFRT-Topas/vhee/vhee_bestopt/results/PDD150MV.npy')
pvdr150MV=np.load('D:Desktop/SFRT-Topas/vhee/vhee_bestopt/results/pvdr150MV.npy')
peak150MV=np.load('D:Desktop/SFRT-Topas/vhee/vhee_bestopt/results/peak150MV.npy')

PDD200MV=np.load('D:Desktop/SFRT-Topas/vhee/vhee_bestopt/results/PDD200MV.npy')
pvdr200MV=np.load('D:Desktop/SFRT-Topas/vhee/vhee_bestopt/results/pvdr200MV.npy')
peak200MV=np.load('D:Desktop/SFRT-Topas/vhee/vhee_bestopt/results/peak200MV.npy')

depths=np.arange(0.25,10,0.5)

figure(1,figsize=(8,8))
plt.errorbar(depths,peak100MV[0],yerr=peak100MV[1],marker='o',markersize=9,capsize=5,capthick=2,label='100 MV')
plt.errorbar(depths,peak150MV[0],yerr=peak150MV[1],marker='o',markersize=9,capsize=5,capthick=2,label='150 MV')
plt.errorbar(depths,peak200MV[0],yerr=peak200MV[1],marker='o',markersize=9,capsize=5,capthick=2,label='200 MV')
plt.ylabel('Mean peak dose rate (Gy/s)')
plt.xlabel('Depth (cm)')
plt.grid(c='lightgray',axis='both',ls='--')
plt.ylim(bottom=0)
plt.xlim(0,10)
plt.legend()

figure(2,figsize=(8,8))
plt.errorbar(depths,pvdr100MV[0],yerr=pvdr100MV[1],marker='o',markersize=9,capsize=5,capthick=2,label='100 MV')
plt.errorbar(depths,pvdr150MV[0],yerr=pvdr150MV[1],marker='o',markersize=9,capsize=5,capthick=2,label='150 MV')
plt.errorbar(depths,pvdr200MV[0],yerr=pvdr200MV[1],marker='o',markersize=9,capsize=5,capthick=2,label='200 MV')
plt.ylabel('PVDR')
plt.xlabel('Depth (cm)')
plt.grid(c='lightgray',axis='both',ls='--')
plt.xlim(0,10)
plt.ylim(bottom=0)
plt.legend()

f = interp1d(pvdr100MV[0], depths)
print('Depth of convergence 100MV: '+str(f(1))+' cm')
f2 = interp1d(pvdr150MV[0], depths)
print('Depth of convergence 150MV: '+str(f2(1))+' cm')
f3 = interp1d(pvdr200MV[0], depths)
print('Depth of convergence 200MV: '+str(f3(1))+' cm')
```
