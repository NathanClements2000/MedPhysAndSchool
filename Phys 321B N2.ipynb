```python
import matplotlib
from matplotlib import pyplot as plt
from matplotlib.pyplot import figure
import numpy as np
from scipy.special import eval_legendre

# For a particle under the influence of gravity in 1D: z(t)=1/2*g*t^2
tstep=0.01
t=np.arange(0.0,2,tstep)

#define z(t)
z=9.81*t-0.5*9.81*t**2

#define function f(t) (a) fn=sin(n*pi/2*t)
f_n=[]
for i in np.arange(1,15):
    f_n.append(np.sin(i*t*np.pi/2))

#Calculate z_n coefficients
z_n=[]
for i in np.arange(0,len(f_n)):
    integral=np.sum(z*f_n[i]*tstep)/(np.sum(f_n[i]**2*tstep))
    z_n.append(integral)
    
z_approx=sum([z_n[i]*f_n[i] for i in np.arange(0,len(f_n))])


figure(1,figsize=(8,8))
plt.plot(t,z,label='z(t)')
plt.plot(t,z_approx,label='approximation of z(t)',ls=':',lw=2)
plt.legend()
plt.ylabel('z (m)')
plt.xlabel('t (s)')
plt.title('(a) f_n(t)=sin(n*t*pi/2)')

print('a) z_n coefficients:'+str(z_n))
print(len(z_n))

#(b) Lets try legendre polynomials as our functions

#define function f(t) (b) f(t)= L_n(t-1)
f_n=[]
for i in np.arange(0,15):
    f_n.append(eval_legendre(i,t-1))

#Calculate z_n coefficients
z_n=[]
for i in np.arange(0,len(f_n)):
    integral=np.sum(z*f_n[i]*tstep)/(np.sum(f_n[i]**2*tstep))
    z_n.append(integral)
    
z_approx=sum([z_n[i]*f_n[i] for i in np.arange(0,len(f_n))])


figure(2,figsize=(8,8))
plt.plot(t,z,label='z(t)')
plt.plot(t,z_approx,label='approximation of z(t)',ls=':',lw=2)
plt.legend()
plt.ylabel('z (m)')
plt.xlabel('t (s)')
plt.title('(b) f_n(t)=L_n(t-1)')

print('b) z_n coefficients:'+str(z_n))


#(c) Try t^n as function this time
#define function f(t) (b) f(t)= t^n
f_n=[]
for i in np.arange(0,15):
    f_n.append(t**i)

#Calculate z_n coefficients
z_n=[]
for i in np.arange(0,len(f_n)):
    integral=np.sum(z*f_n[i]*tstep)/(np.sum(f_n[i]**2*tstep))
    z_n.append(integral)
    
z_approx=sum([z_n[i]*f_n[i] for i in np.arange(0,len(f_n))])

figure(3,figsize=(8,8))
plt.plot(t,z,label='z(t)')
plt.plot(t,z_approx,label='approximation of z(t)',ls=':',lw=2)
plt.legend()
plt.ylabel('z (m)')
plt.xlabel('t (s)')
plt.title('(c) f_n(t)=t^n')

print('c) z_n coefficients:'+str(z_n))
print(len(z_n))

print('We see that we have very nice overlap for the first two functions \
but using t^n as our f_n(t)=t^n to approximate z(t) did not work well, this is because \
t^n (for pos int n) can only have positive first and second derivatives which does not allow\
it to accurately conform to the known z(t)')
```
