```python
import numpy as np
import scipy as sp
from scipy.integrate import odeint
from matplotlib import pyplot as plt
from matplotlib.pyplot import figure

# Lets start with cartesian

#define initial conditions
x0=1
y0=2
vx0=1
vy0=0
a=2.5
m=1
initconds=[x0,vx0,y0,vy0]
#define ODE using our Eqns of motion
def model(y,t,m,a):
    x1,vx1,y1,vy1=y
    dydt= [vx1,-1/m*a*x1*(x1**2+y1**2)**(a/2-1),vy1,-1/m*a*y1*(x1**2+y1**2)**(a/2-1)]
    return dydt
t=np.arange(0,10,0.01) #define a time interval for the solution
sol = odeint(model, initconds, t, args=(m,a)) #use odeint to solve
#unpack the solutions for x,y,vx, and vy
xpos= sol[:,0] 
vx= sol[:,1]
ypos= sol[:,2]
vy=sol[:,3]


#plot trajectory
figure(1,figsize=(8,8))
plt.plot(xpos,ypos)
```

```python
#Lets try polar now

#define initial conditions
r0=np.sqrt(x0^2+y0**2)
theta0=np.arctan(y0/x0)
vr0=1
vtheta0=0
a=2.5
m=1
initconds=[r0,vr0,theta0,vtheta0]
#define ODE using our Eqns of motion
def model2(y,t,m,a):
    r1,vr1,theta1,vtheta1=y
    dydt= [vr1,-1/m*a*r1**(a-1)+r1*vtheta1**2,vtheta1,-2*vr1*vtheta1/r1] #d(theta')/dt defined by initial conditions
    return dydt
t=np.arange(0,10,0.01) #define a time interval for the solution
sol = odeint(model2, initconds, t, args=(m,a)) #use odeint to solve
#unpack the solutions for r and theta
rsol= sol[:,0] 
thetasol= sol[:,2]

#plot trajectory
figure(2,figsize=(8,8))
plt.plot(rsol*np.cos(thetasol),rsol*np.sin(thetasol))

print('Description of what I did: I found the Langrangian and determined the equations of motion using the Euler-Lagrange\
 equation for cartesian and polar coordinates. I then solved the equations of motion for the displacements using\
 the odeint function from scipy and plotted the trajectories using matplotlib.')
```
