```python
import numpy as np
import scipy as sp
from scipy.integrate import odeint
from matplotlib import pyplot as plt
from matplotlib.pyplot import figure
from scipy import integrate

#4c
#define values in SI units
mu_0=4*np.pi*10**-7
a=10*10**-2;b=30*10**-2;k=10*10**-2;I=2;R=0.5;v=3; w=2*np.pi*v

#define time and equation for power(t)
tstep=0.001
t=np.arange(0,np.pi/w,tstep)
def power(t):
    return mu_0**2*I**2*a**2/(4*np.pi**2*R)*(16*b**2*k**2*w**2*(np.cos(w*t))**2/(4*k**2*(np.sin(w*t))**2-b**2)**2)

figure(1,figsize=(8,8))
plt.plot(t,power(t))
plt.grid()
plt.title('Power(t)')
plt.xlabel('Time (s)')
plt.ylabel('Power (W)')

#this is the curve we need to integrate
Etot=sum(power(t)*tstep)
print('i)Etot in one cycle of wire motion is approx '+str(Etot)+' J')

#attempt at ii)
#By cons of E, the force on the wire should be the energy/(wire displacement) aka (integral of our power(t)/ksin(wt))

integrals = []
t_range = []
y_range = []
for i in t:
    t_range.append(i)
    y_range.append(power(i))
    integral = integrate.simps(y_range, t_range)
    integrals.append(integral)

figure(2,figsize=(8,8))
plt.plot(t, integrals)
plt.grid()
plt.title('E(t)')
plt.xlabel('Time (s)')
plt.ylabel('Energy (J)')

Force=power(t)/k*np.sin(w*t)
figure(3,figsize=(8,8))
plt.plot(t, Force)
plt.grid()
plt.title('F(t)')
plt.xlabel('Time (s)')
plt.ylabel('Force (N)')
```
