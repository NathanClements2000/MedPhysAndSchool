```python
import numpy as np
import matplotlib.pyplot as plt
from numpy import genfromtxt
from matplotlib.pyplot import figure
from scipy.signal import savgol_filter


noc=1024 #number of counts
counts = genfromtxt("Downloads/150_with_filter.mca", skip_header=16,skip_footer=58)


sfvalues = genfromtxt("Downloads/150_with_filter.mca",skip_header=14,max_rows=1) #scaling factors
sf= sfvalues[1]/sfvalues[0]
energy = np.linspace(sf,noc*sf,noc)

figure(figsize=(8, 4), dpi=100)
plt.plot(energy,counts)
plt.xlabel('Energy (keV)')
plt.ylabel('Counts')
plt.title('Original')
plt.xlim(0,150)

figure(2)
figure(figsize=(8, 4), dpi=100)
scounts = savgol_filter(counts, 31, 11)
plt.plot(energy,scounts)
plt.xlabel('Energy (keV)')
plt.ylabel('Counts')
plt.title('Smoothened')
plt.xlim(0,150)
```
