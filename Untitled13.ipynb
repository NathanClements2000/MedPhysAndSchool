```python
import numpy as np
import xml.etree.ElementTree as ET
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure

#0.05 xml file
energy1=[]
energydep1=[]
file1=ET.parse('Desktop/EnergyDepInCdTeCrystalHist120keV5x5 0.05.xml')
for btype in file1.findall('./histogram1d/data1d/bin1d'):
        energy1 = np.append(energy1, float(btype.get('binNum')))
        energydep1 = np.append(energydep1, float(btype.get('entries')))
figure(1,figsize=(10,10))
#plt.plot(energy1,energydep1,label='0.05 mm')
plt.yscale('log')
#0.005 xml file
energy2=[]
energydep2=[]
file2=ET.parse('Desktop/EnergyDepInCdTeCrystalHist120keV5x5 0.005.xml')
for btype in file2.findall('./histogram1d/data1d/bin1d'):
        energy2 = np.append(energy2, float(btype.get('binNum')))
        energydep2 = np.append(energydep2, float(btype.get('entries')))
plt.plot(energy2,energydep2-energydep1)
plt.legend()
```
