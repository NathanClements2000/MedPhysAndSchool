```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure
import matplotlib
import scipy
from scipy.signal import find_peaks
import pandas as pd
matplotlib.rcParams['figure.dpi'] = 100
font = {'family' : 'Arial',
        'weight' : 'normal',
        'size'   : 30}

matplotlib.rc('font', **font)
#currents from 1-25 ohms
I=np.array([135.8,135.6,136,135.7,135.0,132.3,129.4,125.7,122,120.2,119.1,114.6,110.6,106.7,103.0,99.5,96.4,93.3,90.4,87.6,85.2,82.7,80.4,78.2,76.1])
I=I*10**-3
R=np.arange(1,25+1)
P=I**2*R
R_err= 0.5/100
I_err= 1/100
P_err=np.sqrt(4*I_err**2+R_err**2)*P
figure(1,figsize=(8,8))
plt.scatter(R[13],np.max(P),c='red',zorder=1,s=40)
plt.plot(R,P,zorder=-1,lw=3)
plt.fill_between(R, P-P_err, P+P_err,alpha=0.4,zorder=-1)
plt.grid(zorder=-1)
plt.xlim(0,26)
plt.ylim(0,0.175)
plt.xlabel('Resistance ($\Omega$)')
plt.ylabel('Power (W)')
P_max=np.max(P)
print('Internal resistance of solar cell '+str(R[13])+str(' ohms'))
print('Max power: '+str(np.around(P_max,3))+' W +/- '+str(np.around(P_err[13],3))+' W')

#Estimated power to solar cell
#sun output power at surface of earth* cos(angle) in sky*(area of solar panels)
P_sun=1000*np.cos(1.22173)*(5.1/100*9.1/100)

print(P_sun)
print('Efficiency of solar cell:'+str((P_max/P_sun)))
```
