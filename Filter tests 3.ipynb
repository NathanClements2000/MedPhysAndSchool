```python
import numpy as np
import matplotlib.pyplot as plt
from numpy import genfromtxt
from matplotlib.pyplot import figure
from skimage.transform import iradon
import matplotlib
import astra

bin_width=120
bins=np.arange(0,120,bin_width)
print(bins)


airscan_5mmAl= [sum(genfromtxt("Desktop/CT imaging/23-6-2021/5mm_Al_airscan.mca", skip_header=int(round(i*550/81+16)),max_rows=int(round(bin_width*550/81)))) for i in bins]
airscan_7mmAl= [sum(genfromtxt("Desktop/CT imaging/23-6-2021/7mm_Al_airscan.mca", skip_header=int(round(i*550/81+16)),max_rows=int(round(bin_width*550/81)))) for i in bins]

Al_5mm= []
avg_counts=[]
for k in np.arange(0,len(bins)):
    for i in np.arange(1,91):
        counts = genfromtxt("Desktop/CT imaging/23-6-2021/5mm_Al_acq{:02d}.mca".format(i), skip_header=int(round(bins[k]*550/81+16)),max_rows=int(round(bin_width*550/81)))
        avg_counts.append(-np.log(sum(counts)/airscan_5mmAl[k]))
        counts=0
    Al_5mm.append(avg_counts)
    avg_counts=[]
    
Al_7mm= []
for k in np.arange(0,len(bins)):
    for i in np.arange(1,91):
        counts = genfromtxt("Desktop/CT imaging/23-6-2021/7mm_Al_acq{:02d}.mca".format(i), skip_header=int(round(bins[k]*550/81+16)),max_rows=int(round(bin_width*550/81)))
        avg_counts.append(-np.log(sum(counts)/airscan_7mmAl[k]))
        counts=0
    Al_7mm.append(avg_counts)
    avg_counts=[]
    
    
    
x= np.arange(1,91)

figure(1,figsize=(8, 5), dpi=100)
plt.plot(x,Al_5mm[0],label='5mm Al')
plt.plot(x,Al_7mm[0],label='7mm Al')
plt.legend()
plt.xlabel('Acquisition Number')
plt.ylabel('-ln(I/I0)')
plt.title('X ray counts for copper filters')
plt.grid(color='darkgrey',linestyle='--')
plt.locator_params(axis="x", nbins=35)


bin_width=1
bins=np.arange(0,130,bin_width)


#First vial
counts_5mmAl = np.array([sum(genfromtxt("Desktop/CT imaging/23-6-2021/5mm_Al_acq19.mca", skip_header=int(round(bins[i]*550/81+16)),max_rows=int(round(bin_width*550/81))))/int(round(bin_width*550/81)) for i in np.arange(0,len(bins))])
counts_7mmAl= np.array([sum(genfromtxt("Desktop/CT imaging/23-6-2021/7mm_Al_acq19.mca", skip_header=int(round(bins[i]*550/81+16)),max_rows=int(round(bin_width*550/81))))/int(round(bin_width*550/81)) for i in np.arange(0,len(bins))])


figure(2,figsize=(10,8))
plt.plot(bins+bin_width/2,counts_5mmAl,label='5mm Al')
plt.plot(bins+bin_width/2,counts_7mmAl,label='7mm Al')
plt.title('PbI2 vial spectrum')
plt.grid(color='darkgrey',linestyle='--')
plt.xlabel('Energy (keV)')
plt.ylabel('Counts')
plt.legend()

#Second vial
counts_5mmAl = np.array([sum(genfromtxt("Desktop/CT imaging/23-6-2021/5mm_Al_acq34.mca", skip_header=int(round(bins[i]*550/81+16)),max_rows=int(round(bin_width*550/81))))/bin_width for i in np.arange(0,len(bins))])
counts_7mmAl = np.array([sum(genfromtxt("Desktop/CT imaging/23-6-2021/7mm_Al_acq34.mca", skip_header=int(round(bins[i]*550/81+16)),max_rows=int(round(bin_width*550/81))))/bin_width for i in np.arange(0,len(bins))])


figure(3,figsize=(10,8))
plt.plot(bins+bin_width/2,(counts_5mmAl),label='5mm Al')
plt.plot(bins+bin_width/2,(counts_7mmAl),label='7mm Al')
plt.title('Au vial spectrum')
plt.grid(color='darkgrey',linestyle='--')
plt.xlabel('Energy (keV)')
plt.ylabel('Counts')
plt.legend()

#Third vial
counts_5mmAl = np.array([sum(genfromtxt("Desktop/CT imaging/23-6-2021/5mm_Al_acq60.mca", skip_header=int(round(bins[i]*550/81+16)),max_rows=int(round(bin_width*550/81))))/bin_width for i in np.arange(0,len(bins))])
counts_7mmAl = np.array([sum(genfromtxt("Desktop/CT imaging/23-6-2021/7mm_Al_acq60.mca", skip_header=int(round(bins[i]*550/81+16)),max_rows=int(round(bin_width*550/81))))/bin_width for i in np.arange(0,len(bins))])


figure(4,figsize=(10,8))
plt.plot(bins+bin_width/2,(counts_5mmAl),label='5mm Al')
plt.plot(bins+bin_width/2,(counts_7mmAl),label='7mm Al')
plt.title('Gd vial spectrum')
plt.grid(color='darkgrey',linestyle='--')
plt.xlabel('Energy (keV)')
plt.ylabel('Counts')
plt.legend()

#Fourth vial
counts_5mmAl = np.array([sum(genfromtxt("Desktop/CT imaging/23-6-2021/5mm_Al_acq73.mca", skip_header=int(round(bins[i]*550/81+16)),max_rows=int(round(bin_width*550/81))))/bin_width for i in np.arange(0,len(bins))])
counts_7mmAl = np.array([sum(genfromtxt("Desktop/CT imaging/23-6-2021/7mm_Al_acq73.mca", skip_header=int(round(bins[i]*550/81+16)),max_rows=int(round(bin_width*550/81))))/bin_width for i in np.arange(0,len(bins))])


figure(5,figsize=(10,8))
plt.plot(bins+bin_width/2,(counts_5mmAl),label='5mm Al')
plt.plot(bins+bin_width/2,(counts_7mmAl),label='7mm Al')
plt.title('I vial spectrum')
plt.grid(color='darkgrey',linestyle='--')
plt.xlabel('Energy (keV)')
plt.ylabel('Counts')
plt.legend()
```
