```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
def W(N):
    Narray=np.arange(1,N)
    ans=0
    for i in Narray:
        ans+= 1/(np.sin(i*np.pi/N)) 
    return (1/N)*ans
Narray= np.arange(2,25)
Warray= [W(i)for i in Narray]
def fit(N,a,b,c):
    return a*((np.log(N))**b)+c
a_fit,b_fit,c_fit= curve_fit(fit,Narray,Warray)[0]
plt.plot(Narray,Warray,label='Work')
plt.xlabel('$N$', fontsize=14)
plt.ylabel(r'$\frac{16\pi\epsilon_0 R}{Q^2}W$', fontsize=14)
plt.title('Work for N=2 to N=25')
plt.grid()
plt.legend()
plt.show()

plt.plot(Narray,Warray,label='Work')
plt.plot(Narray,fit(Narray,a_fit,b_fit,c_fit),label=r'Fitted curve $a (\ln(N))^b+c$',marker='.',linestyle='')
plt.xlabel('$N$', fontsize=14)
plt.ylabel(r'$\frac{16\pi\epsilon_0 R}{Q^2}W$', fontsize=14)
plt.title('Work for N=2 to N=25')
plt.grid()
plt.legend()
plt.show()
print('For the fit we find a, b, and c, respectively')
print(a_fit,b_fit,c_fit)

```
