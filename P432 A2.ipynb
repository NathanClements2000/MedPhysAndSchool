```python
import numpy as np
import matplotlib.pyplot as plt
from numpy import genfromtxt
from matplotlib.pyplot import figure
import matplotlib
%matplotlib inline
from numpy import genfromtxt

# I copied the nist data to notepad files and read them in here
mac_air=genfromtxt("D:Desktop/air_attenuation.txt",skip_header=1).T
mac_water=genfromtxt("D:Desktop/water_attenuation.txt",skip_header=1).T
mac_bone=genfromtxt("D:Desktop/bone_attenuation.txt",skip_header=1).T

#a
figure(1,figsize=(8,8))
plt.plot(mac_air[0],mac_water[2]/mac_air[2])
plt.title('Water to Air ')
plt.xlabel('Energy (MeV)')
plt.ylabel('Ratio of mass attenuation coefficients (Water/Air)')

#b
figure(2,figsize=(8,8))
plt.plot(mac_air[0],mac_bone[2]/mac_water[2])
plt.title('Bone to Water')
plt.xlabel('Energy (MeV)')
plt.ylabel('Ratio of mass attenuation coefficients (Bone/Water)')

#c
maxandmin=[]
for i in np.arange(0,len(mac_air[0])):
    if (mac_air[2][i]/mac_bone[2][i] <=1.8) & (mac_air[2][i]/mac_bone[2][i] >=0.2):
        maxandmin.append(mac_air[0][i])
print('The energy range with better than 20% accuracy would be ' +str(min(maxandmin))+' MeV to '+str(max(maxandmin))+' MeV')
        
        
```
