```python
import matplotlib.pyplot as plt
import numpy as np
a_0= 1
a1=0
a2= np.pi/6
a3=np.pi/4
a4=np.pi/3
a5=np.pi/2

def st200(x,y,z):
    r=np.sqrt(x**2+y**2+z**2)
    R20=1/np.sqrt(2)*(1-r/2)*np.exp(-r/2)
    Y00=1/np.sqrt(4*np.pi)
    return R20*Y00
def st210(x,y,z):
    r=np.sqrt(x**2+y**2+z**2)
    R21= 1/np.sqrt(24)*r*np.exp(-r/2)
    Y10=np.sqrt(3/(np.pi*4))*z/r
    return R21*Y10
def psi(x,y,z,a):
    return np.sqrt(3/5)*st200(x,y,z)+np.exp(1j*a)*np.sqrt(2/5)*st210(x,y,z)

def prob(x,y,z,a):
    return np.abs(psi(x,y,z,a))**2
y=0
x,z= np.meshgrid(np.linspace(-10,10,500),np.linspace(-10,10,500))
prob1= np.vectorize(prob)(x,y,z,a1)
prob2= np.vectorize(prob)(x,y,z,a2)
prob3= np.vectorize(prob)(x,y,z,a3)
prob4= np.vectorize(prob)(x,y,z,a4)
prob5= np.vectorize(prob)(x,y,z,a5)

plt.imshow(prob1,extent=[-10,10,-10,10],cmap='magma')
plt.colorbar()
plt.clim(0,0.0014)
plt.xlabel('x')
plt.ylabel('z')
plt.title(r"Probability Density for Hydrogen ($\alpha=0$)")
plt.show()

plt.imshow(prob2,extent=[-10,10,-10,10],cmap='magma')
plt.colorbar()
plt.clim(0,0.0014)
plt.xlabel('x')
plt.ylabel('z')
plt.title(r"Probability Density for Hydrogen ($\alpha=\pi/6$)")
plt.show()

plt.imshow(prob3,extent=[-10,10,-10,10],cmap='magma')
plt.colorbar()
plt.clim(0,0.0014)
plt.xlabel('x')
plt.ylabel('z')
plt.title(r"Probability Density for Hydrogen ($\alpha=\pi/4$)")
plt.show()

plt.imshow(prob4,extent=[-10,10,-10,10],cmap='magma')
plt.colorbar()
plt.clim(0,0.0014)
plt.xlabel('x')
plt.ylabel('z')
plt.title(r"Probability Density for Hydrogen ($\alpha=\pi/3$)")
plt.show()

plt.imshow(prob5,extent=[-10,10,-10,10],cmap='magma')
plt.colorbar()
plt.clim(0,0.0014)
plt.xlabel('x')
plt.ylabel('z')
plt.title(r"Probability Density for Hydrogen ($\alpha=\pi/2$)")
plt.show()

print("b) alpha changes the weighting of each two states in their superposition ")
```
