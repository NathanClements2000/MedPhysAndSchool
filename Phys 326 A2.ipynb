```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import quad

def integrandx(phi, x_R, y_R, z_R):
    return np.sin(phi)*(x_R-np.cos(phi))/(((x_R-np.cos(phi))**2+(y_R-np.sin(phi))**2+(z_R)**2)**(3/2))
def electricfieldx(x_R, y_R, z_R):
    return quad(integrandx, 0, 2*np.pi, args=(x_R, y_R, z_R))[0]

x_R = np.linspace(-1/2,1/2,1000)
electricfieldsx = np.vectorize(electricfieldx)(x_R, 1/2, 0)
plt.plot(x_R, electricfieldsx)
plt.xlabel('$x/R$', fontsize=14)
plt.ylabel('aEx', fontsize=14)
plt.grid()
plt.title('Electric field in x direction')
plt.show()

def integrandy(phi, x_R, y_R, z_R):
    return np.sin(phi)*(y_R-np.sin(phi))/(((x_R-np.cos(phi))**2+(y_R-np.sin(phi))**2+(z_R)**2)**(3/2))
def electricfieldy(x_R, y_R, z_R):
    return quad(integrandy, 0, 2*np.pi, args=(x_R, y_R, z_R))[0]

electricfieldsy = np.vectorize(electricfieldy)(x_R, 1/2, 0)
plt.plot(x_R, electricfieldsy)
plt.xlabel('$x/R$', fontsize=14)
plt.ylabel('aEy', fontsize=14)
plt.grid()
plt.title('Electric field in y direction')
plt.show()
```
