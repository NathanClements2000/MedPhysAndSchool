```python
import numpy as np
import pandas as pd
import os
from matplotlib import pyplot as plt
from matplotlib.pyplot import figure

%matplotlib inline
Data = pd.read_csv('d:/Downloads/ML-Mar22.csv')

Times = Data['Data Set 1:Time(s)'].copy()

Freqs = np.linspace(0.1, 1.5, 300)
FCosTrans1 = np.zeros(300)
FSinTrans1 = np.zeros(300)
FCosTrans2 = np.zeros(300)
FSinTrans2 = np.zeros(300)
FCosTrans3 = np.zeros(300)
FSinTrans3 = np.zeros(300)

for i in range(len(Freqs)):
    FourierC = np.cos(2*np.pi*Freqs[i]*Times)
    ConvoluteC = FourierC*Data['Data Set 1:X-axis acceleration 1(m/s)']
    FourierS = np.sin(2*np.pi*Freqs[i]*Times)
    ConvoluteS = FourierS*Data['Data Set 1:X-axis acceleration 1(m/s)']    
    integral = ConvoluteC.sum()*(Times[1]-Times[0])
    FCosTrans1[i] = integral
    integral = ConvoluteS.sum()*(Times[1]-Times[0])
    FSinTrans1[i] = integral
    


for i in range(len(Freqs)):
    FourierC = np.cos(2*np.pi*Freqs[i]*Times)
    ConvoluteC = FourierC*Data['Data Set 1:X-axis acceleration 2(m/s)']
    FourierS = np.sin(2*np.pi*Freqs[i]*Times)
    ConvoluteS = FourierS*Data['Data Set 1:X-axis acceleration 2(m/s)']    
    integral = ConvoluteC.sum()*(Times[1]-Times[0])
    FCosTrans2[i] = integral
    integral = ConvoluteS.sum()*(Times[1]-Times[0])
    FSinTrans2[i] = integral
    

for i in range(len(Freqs)):
    FourierC = np.cos(2*np.pi*Freqs[i]*Times)
    ConvoluteC = FourierC*Data['Data Set 1:X-axis acceleration 3(m/s)']
    FourierS = np.sin(2*np.pi*Freqs[i]*Times)
    ConvoluteS = FourierS*Data['Data Set 1:X-axis acceleration 3(m/s)']    
    integral = ConvoluteC.sum()*(Times[1]-Times[0])
    FCosTrans3[i] = integral
    integral = ConvoluteS.sum()*(Times[1]-Times[0])
    FSinTrans3[i] = integral
    
fourier_m1_run1=np.sqrt(FCosTrans1**2 + FSinTrans1**2)
fourier_m2_run1=np.sqrt(FCosTrans2**2 + FSinTrans2**2)
fourier_m3_run1=np.sqrt(FCosTrans3**2 + FSinTrans3**2)

figure(1)
plt.plot(Freqs, np.sqrt(FCosTrans1**2 + FSinTrans1**2)+2,label='Mass 1')
plt.plot(Freqs, np.sqrt(FCosTrans2**2 + FSinTrans2**2),label = 'Mass 2')
plt.plot(Freqs, np.sqrt(FCosTrans3**2 + FSinTrans3**2)-2, label = 'Mass 3')
plt.legend(loc = 'upper right')
plt.xlabel('Frequency (Hz)')
plt.ylabel('Offset Acceleration Amplitude ($cm/s^2$)')
plt.title('Fourier Spectrum for run 1 - Absolute values')
plt.show()

Freqs = np.linspace(0.1, 1.5, 300)
FCosTrans1 = np.zeros(300)
FSinTrans1 = np.zeros(300)
FCosTrans2 = np.zeros(300)
FSinTrans2 = np.zeros(300)
FCosTrans3 = np.zeros(300)
FSinTrans3 = np.zeros(300)


for i in range(len(Freqs)):
    FourierC = np.cos(2*np.pi*Freqs[i]*Times)
    ConvoluteC = FourierC*Data['Data Set 2:X-axis acceleration 1(m/s)']
    FourierS = np.sin(2*np.pi*Freqs[i]*Times)
    ConvoluteS = FourierS*Data['Data Set 2:X-axis acceleration 1(m/s)']    
    integral = ConvoluteC.sum()*(Times[1]-Times[0])
    FCosTrans1[i] = integral
    integral = ConvoluteS.sum()*(Times[1]-Times[0])
    FSinTrans1[i] = integral
    


for i in range(len(Freqs)):
    FourierC = np.cos(2*np.pi*Freqs[i]*Times)
    ConvoluteC = FourierC*Data['Data Set 2:X-axis acceleration 2(m/s)']
    FourierS = np.sin(2*np.pi*Freqs[i]*Times)
    ConvoluteS = FourierS*Data['Data Set 2:X-axis acceleration 2(m/s)']    
    integral = ConvoluteC.sum()*(Times[1]-Times[0])
    FCosTrans2[i] = integral
    integral = ConvoluteS.sum()*(Times[1]-Times[0])
    FSinTrans2[i] = integral
    

for i in range(len(Freqs)):
    FourierC = np.cos(2*np.pi*Freqs[i]*Times)
    ConvoluteC = FourierC*Data['Data Set 2:X-axis acceleration 3(m/s)']
    FourierS = np.sin(2*np.pi*Freqs[i]*Times)
    ConvoluteS = FourierS*Data['Data Set 2:X-axis acceleration 3(m/s)']    
    integral = ConvoluteC.sum()*(Times[1]-Times[0])
    FCosTrans3[i] = integral
    integral = ConvoluteS.sum()*(Times[1]-Times[0])
    FSinTrans3[i] = integral
    
fourier_m1_run2=np.sqrt(FCosTrans1**2 + FSinTrans1**2)
fourier_m2_run2=np.sqrt(FCosTrans2**2 + FSinTrans2**2)
fourier_m3_run2=np.sqrt(FCosTrans3**2 + FSinTrans3**2)

figure(2)
plt.plot(Freqs, np.sqrt(FCosTrans1**2 + FSinTrans1**2)+2,label='Mass 1')
plt.plot(Freqs, np.sqrt(FCosTrans2**2 + FSinTrans2**2),label = 'Mass 2')
plt.plot(Freqs, np.sqrt(FCosTrans3**2 + FSinTrans3**2)-2, label = 'Mass 3')
plt.legend(loc = 'upper right')
plt.xlabel('Frequency (Hz)')
plt.ylabel('Offset Acceleration Amplitude ($cm/s^2$)')
plt.title('Fourier Spectrum for run 2 - Absolute values')
plt.show()

Freqs = np.linspace(0.1, 1.5, 300)
FCosTrans1 = np.zeros(300)
FSinTrans1 = np.zeros(300)
FCosTrans2 = np.zeros(300)
FSinTrans2 = np.zeros(300)
FCosTrans3 = np.zeros(300)
FSinTrans3 = np.zeros(300)


for i in range(len(Freqs)):
    FourierC = np.cos(2*np.pi*Freqs[i]*Times)
    ConvoluteC = FourierC*Data['Data Set 3:X-axis acceleration 1(m/s)']
    FourierS = np.sin(2*np.pi*Freqs[i]*Times)
    ConvoluteS = FourierS*Data['Data Set 3:X-axis acceleration 1(m/s)']    
    integral = ConvoluteC.sum()*(Times[1]-Times[0])
    FCosTrans1[i] = integral
    integral = ConvoluteS.sum()*(Times[1]-Times[0])
    FSinTrans1[i] = integral
    


for i in range(len(Freqs)):
    FourierC = np.cos(2*np.pi*Freqs[i]*Times)
    ConvoluteC = FourierC*Data['Data Set 3:X-axis acceleration 2(m/s)']
    FourierS = np.sin(2*np.pi*Freqs[i]*Times)
    ConvoluteS = FourierS*Data['Data Set 3:X-axis acceleration 2(m/s)']    
    integral = ConvoluteC.sum()*(Times[1]-Times[0])
    FCosTrans2[i] = integral
    integral = ConvoluteS.sum()*(Times[1]-Times[0])
    FSinTrans2[i] = integral
    

for i in range(len(Freqs)):
    FourierC = np.cos(2*np.pi*Freqs[i]*Times)
    ConvoluteC = FourierC*Data['Data Set 3:X-axis acceleration 3(m/s)']
    FourierS = np.sin(2*np.pi*Freqs[i]*Times)
    ConvoluteS = FourierS*Data['Data Set 3:X-axis acceleration 3(m/s)']    
    integral = ConvoluteC.sum()*(Times[1]-Times[0])
    FCosTrans3[i] = integral
    integral = ConvoluteS.sum()*(Times[1]-Times[0])
    FSinTrans3[i] = integral
    
fourier_m1_run3=np.sqrt(FCosTrans1**2 + FSinTrans1**2)
fourier_m2_run3=np.sqrt(FCosTrans2**2 + FSinTrans2**2)
fourier_m3_run3=np.sqrt(FCosTrans3**2 + FSinTrans3**2)

figure(3)
plt.plot(Freqs, np.sqrt(FCosTrans1**2 + FSinTrans1**2)+2,label='Mass 1')
plt.plot(Freqs, np.sqrt(FCosTrans2**2 + FSinTrans2**2),label = 'Mass 2')
plt.plot(Freqs, np.sqrt(FCosTrans3**2 + FSinTrans3**2)-2, label = 'Mass 3')
plt.legend(loc = 'upper right')
plt.xlabel('Frequency (Hz)')
plt.ylabel('Offset Acceleration Amplitude ($cm/s^2$)')
plt.title('Fourier Spectrum for run 3 - Absolute values')
plt.show()

figure(4)
plt.plot(Freqs,FCosTrans1,label="Cosine series")
plt.plot(Freqs,FSinTrans1,label='Sine series')
plt.xlabel('Frequency (Hz)')
plt.legend()
```
